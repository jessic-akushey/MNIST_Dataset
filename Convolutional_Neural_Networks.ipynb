```python
import pandas as pd
import numpy as np

# TensorFlow and tf.keras
import tensorflow as tf
import keras

from tensorflow.keras import Sequential

from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.losses import KLDivergence

from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
df= pd.read_csv("mnist.csv")
```

```python
df.drop('id', axis = 1, inplace = True)
df.head()
```

```python
df.describe()
```

### Splitting the data into Train and Test set

```python
# Splitting the features into predictors and predictor
X = df.drop('class', axis = 1)
y = df['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print(f'X trainset size {x_train.shape}')
print(f'X testset size {x_test.shape}')
print(f'y trainset size {y_train.shape}')
print(f'y testset size {y_test.shape}')
```

```python
# # Reshaping each row of the train data into a 28 by 28 array with a dimension of 1

x_train = x_train.values.reshape(-1,28,28, 1)
```

```python
x_train.shape
```

The description of the data shows that the maximum value for each column is __255__ and the minimum is __0__. To have the range of values between __0__ and __1__, the values could be simply divided by __255__.

```python
# Scaling the data using the MinMax Scaler
x_train = x_train/255.0
```

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, 10)
```

## Building the Convolutional Model

```python
model = keras.Sequential(
    [
        keras.Input(shape=(28, 28, 1)),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(10, activation="softmax"),
    ]
)

model.summary()
```

```python
batch_size = 100
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

## Predicting using the Test set

```python
x_test = x_test.values.reshape(-1,28,28, 1)

# convert class vectors to binary class matrices
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
